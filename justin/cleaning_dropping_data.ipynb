```python
import pandas as pd
import numpy as np
```

```python
#import the covid deaths csv
deaths_df = pd.read_csv('deaths.csv')
list(deaths_df.columns)
```

```python
deaths_df = deaths_df.drop(['Sex','Total Deaths', 'Pneumonia Deaths', 'Pneumonia and COVID-19 Deaths','Influenza Deaths', 'Pneumonia, Influenza, or COVID-19 Deaths','Footnote', 'Age Group', 'Year', 'Month'], axis=1)
```

```python
deaths_df = deaths_df.loc[(deaths_df['State'] != 'United States')]
deaths_df.head()
```

```python
deaths_df = deaths_df.drop(['Group'], axis = 1)
deaths_df.head()
```

```python
deaths_df = deaths_df.drop(['Data As Of'], axis=1)
```

```python
deaths_df = deaths_df.dropna()
```

```python
deaths_df[['Start Date','End Date']] = deaths_df[['Start Date', 'End Date']].apply(pd.to_datetime)
```

```python
deaths_df['COVID-19 Deaths'] = pd.to_numeric(deaths_df['COVID-19 Deaths'], errors = 'coerce')
```

```python
deaths_df = deaths_df[deaths_df.State != 'Puerto Rico']
```

```python
deaths_df = deaths_df[deaths_df.State != 'District of Columbia']
```

```python
deaths_df.replace(
    to_replace= ['New York City'], 
    value = 'New York',
    inplace= True
)
deaths_df
```

```python
prevax= deaths_df[deaths_df['End Date']<='2021-01-02'].groupby('State').sum('Covid-19 Deaths')
prevax
```

```python
postvax= deaths_df[deaths_df['End Date']>='2021-01-02'].groupby('State').sum('Covid-19 Deaths')
postvax
```

```python
prevax.to_csv('prevax_deaths.csv')
```

```python
postvax.to_csv('postvax_deaths.csv')
```
